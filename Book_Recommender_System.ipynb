```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating': 'num_ratings'}, inplace = True)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns = {'Book-Rating': 'avg_rating'}, inplace = True)
```

```python
popularity_df = pd.merge(num_rating_df, avg_rating_df, on='Book-Title')
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values(by='avg_rating', ascending=False).head(50)
```

```python
popularity_df = popularity_df.merge(books, on='Book-Title').drop_duplicates(subset='Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

Collaborative Filtering Based Recommendation System

```python
Rated_user = ratings_with_name.groupby('User-ID').count()['Book-Rating']>=200
Rated_users = Rated_user[Rated_user].index
```

```python
Filter_rating = ratings_with_name[ratings_with_name['User-ID'].isin(Rated_users)]
```

```python
Famous_book = Filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
Famous_books = Famous_book[Famous_book].index
```

```python
Final_ratings = Filter_rating[Filter_rating['Book-Title'].isin(Famous_books)]
```

```python
Final_ratings
```

```python
pt = Final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score = cosine_similarity(pt)
```

```python
Similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(Similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("1984")
```
